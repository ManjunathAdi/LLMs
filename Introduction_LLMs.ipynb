# Phi-3

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load model and tokenizer
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code=True,
)
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")
```

```python
from transformers import pipeline

# Create a pipeline
generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    return_full_text=False,
    max_new_tokens=800,
    do_sample=False
)
```

```python
# The prompt (user input / query)
messages = [
    {"role": "user", "content": "Create a funny joke about kapil sharma show"}
]

# Generate output
output = generator(messages)
print(output[0]["generated_text"])
```

```python
prompt = "Need brief summarization of vedas and upanishads. <|assistant|>"

# Tokenize the input prompt
input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to("cuda")

# Generate the text
generation_output = model.generate(
  input_ids=input_ids,
  max_new_tokens=800,
  do_sample=True,
  top_p=0.9,
  temperature=1.0,
  repetition_penalty=1.2
)

# Print the output
print(tokenizer.decode(generation_output[0]))
```

# Other Tokenizers

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

def show_tokens(sentence, tokenizer_name):
    tokenizer = AutoTokenizer.from_pretrained(tokenizer_name)
    token_ids = tokenizer(sentence).input_ids
    for idx, t in enumerate(token_ids):
        print(
            tokenizer.decode(t) +
            '\x1b[0m',
            end=' '
        )
```

```python
text = """
India and NEWDELHI
🎵 鸟
display the tokens False None elif == >= else:  Three tabs: "       "
6.0*80=480
"""
```

# Input-Output of trained LLM: Phi-3

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

# Load model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")

model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code=True,
)

# Create a pipeline
generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    return_full_text=False,
    max_new_tokens=60,
    do_sample=False,
)
```

```python
prompt = "Explain the horrific killing of Hindus done by muslim kings in India. Explain how it happened."

output = generator(prompt)

print(output[0]['generated_text'])
```

# Selecting single token with max probabaility

```python

prompt = "The capital of Japan is"

# Tokenize the input prompt
input_ids = tokenizer(prompt, return_tensors="pt").input_ids

# Tokenize the input prompt
input_ids = input_ids.to("cuda")

# Get the output of the model before the lm_head
model_output = model.model(input_ids)

# Get the output of the lm_head
lm_head_output = model.lm_head(model_output[0])
```

```python
lm_head_output.shape
```

```python
lm_head_output[0,-1].argmax(-1)
```

```python
token_id = lm_head_output[0,-1].argmax(-1)
tokenizer.decode(token_id)
```

# Usage of use_cache=True/False

```python
prompt = "Write a very long essay on the topic of greatest tennis player of all time"

# Tokenize the input prompt
input_ids = tokenizer(prompt, return_tensors="pt").input_ids
input_ids = input_ids.to("cuda")
```

```python
%%timeit -n 1
# Generate the text
generation_output = model.generate(
  input_ids=input_ids,
  max_new_tokens=80,
  use_cache=True
)
```

```python
%%timeit -n 1
# Generate the text
generation_output = model.generate(
  input_ids=input_ids,
  max_new_tokens=80,
  use_cache=False
)
```
